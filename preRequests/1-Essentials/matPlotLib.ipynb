```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
yasArr=np.array([10,20,30,40,50,60,70,80,90])
```

```python
kiloArr=np.random.randint(20,180,9)
# x and y must have same first dimension
```

```python
plt.plot(yasArr,kiloArr,"g*-")
# plt.plot (x ekseni,y eksseni,"color")
plt.xlabel("Yas")
plt.ylabel("Kilo")
plt.title("yas ile kilo grafigi")
## plt.show kullanmiyoz cunku %matplotlib inline kullandik
```

## Ozellestirme

```python
npd1=np.linspace(0,40,9)
# 0 ile 40 arasinda 10 tane eleman olustur 
# 40 dahil (stopValur dahil linspace 'de)
# linSpace start ve stop values dahil edecek sekilde araliklari ayarliyor 
```

```python
npd2 =npd1**3
```

```python
npd1
```

```python
npd2
```

```python
plt.plot(npd1,npd2,"r")
```

```python
plt.plot(npd2,npd1,"y*-") #  yukarda'kinin tersi 
# burada her yildiz bir veri temsil ediyor bizim icin
# veri noktalari nerede durdugunu gormek istiyorsak icin kullanilir
```

* **ayni duzlem'de iki tane farkli grafik olusturma**

```python
plt.subplot(1,2,2)
## (sira sayisi ,sutun sayisi ,hangisini cizecek)
plt.plot(npd1,npd2,"r")
plt.subplot(1,2,1)
plt.plot(npd2,npd1,"y*-") 
## birbirine cok yapistini  gordukl bu yuzden sunu kullaniriz 
plt.tight_layout()
```

## ic_ice_grafikler  & Figure

```python
## bunlar ayni satir'da calistirilmasi lazim
my_figure=plt.figure()
# figure 'ler bos olur  axes =0
#o yuzden axes eklemem gerekiyor 
my_axes =my_figure.add_axes([0.1,0.1,0.4,0.9])
#add([left ,bottom,weight ,height])
my_axes.plot(npd1,npd2,"b")
my_axes.set_xlabel("npd1")
# cizilen sey axes'tir figure degil'dir
```

```python
figure2=plt.figure()
axes1=figure2.add_axes([0.1,0.1,0.9,0.9])# buarada sadece box'un attributeleri
axes2=figure2.add_axes([0.3,0.5,0.3,0.3])
axes1.plot(npd1,npd2,"b")
axes2.plot(npd2,npd1,"r")
```

## subPlots

```python
#*plt.subplots()
#bu hem figure hemde axes dondurur =
##ve bunlar Tuple olarak donduruyor
### dondurdugu Tuple bu sekil'de esitleyebiliriz 
(figurem,axesim)=plt.subplots()
axesim.plot(npd1,npd2,"g")
```

```python
(figurem,axesim)=plt.subplots(1,2)
## subplot olusturdugu axes numpy.ndarray turunden
#axesim.plot(npd1,npd2,"g")
for ax in axesim:
    ax.plot(npd1,npd2,"g")
plt.tight_layout()
```

 ## gorseller_gelistirme

```python
yeniFigure =plt.figure()
yeniAx=yeniFigure.add_axes([0.1,0.1,0.9,0.9])
yeniAx.plot(npd1,npd2,label="npd1**3")
plt.legend(loc=6)
```

```python
yeniFigure =plt.figure(dpi=150)## figure icine argument verebiliriz 
#dpi = kac pixel  falan ile ilgili  grafigin kalitesi
yeniAx=yeniFigure.add_axes([0.1,0.1,0.9,0.9])
yeniAx.plot(npd1,npd2,label="npd1**3")
yeniAx.plot(npd1,npd1**5,label="npd1**5")
plt.legend(loc=2)
```

```python
yeniFigure =plt.figure()## figure icine argument verebiliriz 
#dpi = kac pixel  falan ile ilgili  grafigin kalitesi
yeniAx=yeniFigure.add_axes([0.1,0.1,0.9,0.9])
yeniAx.plot(npd1,npd2,label="npd1**3")
yeniAx.plot(npd1,npd1**5,label="npd1**5",color="000000")
plt.legend(loc=2)
```

```python
# dpi grafigi kaydetmek istedigimiz de onun kalitesini
##arttirarak dpi kullanarak kaydediyoruz 
yeniFigure.savefig("benimfigure.png",dpi=250)
## jupyter_noteBook nerede calisiyorsa orada kaydediliyor
```

## gorsellik_ileri_seviye

```python
(fig,axe)=plt.subplots()
#<matplotlib.lines.Line2D at 0x7f670511cfd0>]
## boyle bir hata gordugun zaman demekki
### plot figure ile ayni cell koyman lazim
axe.plot(npd1,npd2,label ="np1**3",
         color ="#000000",linewidth=3.0,linestyle="--",marker ="o",markerfacecolor="red",markersize=9.0)
#marker verilerin nerede oldugunu belirler  ve ici doldurulabillir bi karakter olmasi bizim icin daha avantajli 
##bir durum yaratiyor
axe.plot(npd1,npd1**5,label ="np1**5",
         color ="#FFFF00",linewidth=3.0,alpha=0.7)
plt.legend()
```
