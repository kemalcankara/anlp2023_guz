```python
import pandas as pd 

df = pd.read_csv("./data/salaries.csv")
df
```

```python
## as we learn early that machine learning models don't work with text so we need to convert texts into numeric values 

from sklearn.preprocessing import LabelEncoder

# for each label we need to create a label encoder object so 
companyLe = LabelEncoder()
jobLe = LabelEncoder()
degreeLe = LabelEncoder()
# then we create new columns inside the data frame 
df["new_company"] = companyLe.fit_transform(df["company"])
df["new_job"] = jobLe.fit_transform(df["job"])
df["new_degree"] = degreeLe.fit_transform(df["degree"])
# then delete old ones 
df.drop(["company","job","degree"],axis=1,inplace=True)

df
```

```python
## now we need to split our data 
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df.drop(["salary_more_then_100k"],axis=1),df["salary_more_then_100k"],test_size=0.2)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier(criterion="entropy")
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
